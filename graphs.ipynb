```python
#OUVERTURE CSV COSMETICS

db_cosmetics = pd.read_csv('./dataset_cosmetics/company_sales_data.csv', encoding='utf-8',  sep=',')
db_cosmetics
```

```python
# GRAPHE 2

ax = plt.gca()

db_cosmetics.plot(kind='line',x='month_number',y='total_profit',ax=ax)
plt.ylabel('Profit')
plt.xlabel('Mois')
plt.show()
```

```python
# GRAPHE 3

# initialisation des donnes
x = np.arange(0,11)
y1 = db_cosmetics.iloc[0:11]['facecream']
y2 = db_cosmetics.iloc[0:11]['facewash']
y3 = db_cosmetics.iloc[0:11]['toothpaste']
y4 = db_cosmetics.iloc[0:11]['bathingsoap']
y5 = db_cosmetics.iloc[0:11]['shampoo']
y6 = db_cosmetics.iloc[0:11]['moisturizer']

# initialisation des axes et du graphe
fig, ax = plt.subplots(1, figsize=(8, 6))
fig.suptitle('Profits ralis pour chaque produit', fontsize=15)
plt.ylabel('Profit')
plt.xlabel('Mois')

# Assignation des labels pour chaque plot
ax.plot(x, y1, label="Faceream")
ax.plot(x, y2, label="Facewash")
ax.plot(x, y3, label="Toothpaste")
ax.plot(x, y4, label="Bathingsoap")
ax.plot(x, y5, label="Shampoo")
ax.plot(x, y6, label="Moisturizer")

# Ajout d'une lgende
plt.legend(loc="upper left", title="Produits", frameon=True)

# Affichage du rsultat
plt.show()
```

```python
# GRAPHE 4

# largeur des barres
barWidth = 0.15
fig = plt.subplots(figsize =(19, 8))
 
# Positionner les barres sur l'axe X
br1 = np.arange(len(y1))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
br4 = [x + barWidth for x in br3]
br5 = [x + barWidth for x in br4]
br6 = [x + barWidth for x in br5]

# Cration des plots
plt.bar(br1, y1, width = barWidth, edgecolor ='grey', label ='Facecream')
plt.bar(br2, y2, width = barWidth, edgecolor ='grey', label ='Facewash')
plt.bar(br3, y3, width = barWidth, edgecolor ='grey', label ='Toothpaste')
plt.bar(br4, y4, width = barWidth, edgecolor ='grey', label ='Bathing')
plt.bar(br5, y5, width = barWidth, edgecolor ='grey', label ='Shampoo')
plt.bar(br6, y6, width = barWidth, edgecolor ='grey', label ='Moist')
 
# Ajouter Xticks
plt.xlabel('Mois', fontweight ='bold', fontsize = 15)
plt.ylabel('Nombres units vendues', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(db_cosmetics['month_number']))], ['Janvier', 'Fevrier', 'Mars', 'Avril', 'Mai', 'Juin', 'Juillet', 'Aout', 'Sept', 'Oct', 'Nov', 'Dec'])

# Affichage du graphe
plt.legend()
plt.show()
```

```python
# GRAPHE 5

# dfinir labels et donnes du graphe
labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
pie = [db_cosmetics['facecream'].sum(), db_cosmetics['facewash'].sum(), db_cosmetics['toothpaste'].sum(), 
         db_cosmetics['bathingsoap'].sum(), db_cosmetics['shampoo'].sum(), db_cosmetics['moisturizer'].sum()]

# Cration graph et lgende
plt.axis("equal")
plt.pie(pie, labels=labels, autopct='%1.1f%%')
plt.legend(loc='lower right')
plt.title('Sales data')

# Affichage graphe
plt.show()
```
